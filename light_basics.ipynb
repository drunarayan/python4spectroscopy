$c = \lambda f$ OR $\lambda = {{c}\over{f}}$

where $c$ is the *speed* of the wave in $m\over{s}$, 

$\lambda$ is *wavelength* in $m$, and 

$f$ is in ${hz}$ which has units of $1\over{s}$

<div class="alert alert-info">üîµ What is the wavelength of visible light that has a frequency of  5.66√ó10^14 Hz
 ? Provide your answer by creating variables and calculating below?</div>

```python
f = 5.66e14 #hz
c = 3e8 #m/s
w = c/f
result = f'The wavelength of visible light that has a frequency of 5.66√ó10^14 Hz is {w:1.2e} m'
print(result)
```

If a tsunami travels at the speed of 600 km/h (speed of an airliner) and approaches a shore at a rate of one wave crest every 15 min (4 waves/h), what would be the distance between those wave crests at sea in km?</div>

```python
c = 600e3/3600 #m/s
f = 4/3600 #hz
w = c/f
result = f'the distance between those wave crests at sea is {w/1000:.0f} km'
print(result)
```
